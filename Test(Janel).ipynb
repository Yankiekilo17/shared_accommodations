```python
# Dependencies
import pandas as pd
import numpy as np
import csv
import requests
import json
from config import api_key
```

```python
df = pd.read_csv("Data/yearly_rent.csv")
df.head()
```

```python
df["Locality"] = ""
df
```

```python
# create a params 
params = {"key": "api_key",
         "types": "administrative_area_level_2"}

# Loop through the year_rent_df and run a lat/long search for each neighborhood
for index, row in df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    neighborhood = row["Neighborhood"]
    city = row["City"]

    # update address key value
    params['address'] = f"{neighborhood},{city}"

    # make request
    locality = requests.get(base_url, params=params)
    
    
    # convert to json
    locality = locality.json()
    
    #inserting coordinates to assigned columns
    df.loc[index, "Locality"] = locality["results"][0]['address_components'][0]['long_name']
   # df.loc[index, "Lng"] = lat_lng["results"][0]["type"]["location"]["lng"]
```
