```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

%matplotlib notebook
```

```python
csv_path = "hp_and_rent_grouped.csv"

merged_df = pd.read_csv(csv_path, encoding='utf8', engine='python')

merged_df.head()
```

```python
neighborhood = merged_df['Neighborhood']
price = merged_df['Assessed Land Value']
rent = merged_df['Avg Rent Rate']
```

```python
N = 5
priceMeans = price
rentMeans = rent
neigh = neighborhood

# Set x and y limits
plt.xlim(-0.25, len(neigh))
plt.ylim(0, max(price)+10)

#objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(neigh))
performance = price
 
plt.bar(y_pos, performance, align='center', alpha=.5)
plt.xticks(y_pos, neigh, rotation = 'vertical')
#plt.get_xticklabels(xlabels, rotation=90)
plt.ylabel('Price')
plt.title('Housing Price')
plt.tight_layout()
plt.show()
```

```python
multi_df = {}
multi_df["Neighborhood"] = merged_df['Neighborhood']
multi_df["Price"] = merged_df['Assessed Land Value']
multi_df["Rent"] = merged_df['Avg Rent Rate']
```

```python
multi_df = pd.DataFrame(multi_df)
multi_df["Price"] = multi_df["Price"] / 5
multi_df.head()
```

```python
# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = multi_df.plot(kind="bar", figsize=(20,5))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(multi_df["Neighborhood"], rotation=90)

plt.show()
plt.tight_layout()
plt.savefig('HP_Rent.png')
```

```python
year_df['Pct Change'] = year_df['Pct Change'] * 10000
```

```python
year_df.head()
```

```python
prev_neigh = 'None'
for index, neigh in year_df.iterrows():
     if neigh["Neighborhood"] != prev_neigh:
        process = year_df.loc[year_df["Neighborhood"] == neigh["Neighborhood"]]
        p_years=[]
        p_amt=[]
        r_amt=[]
        c_pct=[]
        for index, county in process.iterrows():
            p_years.append(county["Year"])
            p_amt.append(county['Assessed Land Value'])
            r_amt.append(county['Yearly Rent Price'])
            c_pct.append(county['Pct Change'])
        price, = plt.plot(p_years, p_amt, color="blue", label= neigh["Neighborhood"] + " Price Average")
        rent, = plt.plot(p_years, r_amt, color="red", label=neigh["Neighborhood"] + " Rent Average")
        per, = plt.plot(p_years, c_pct, color="green", label=neigh["Neighborhood"] + " Pct Change")
        # # Create a legend for our chart

        # Create a legend for our chart
        plt.legend(handles=[price, rent, per], loc="best")
        # Show the chart
        plt.show()    
        prev_neigh = neigh["Neighborhood"]
```
